一人当たりのGDPが高いほど寄付などがふえるのではないかという仮説が正しいかを見ていきます。

```python
# 必要なライブラリをインポート
import pandas as pd
import matplotlib.pyplot as plt

# CSVファイルを読み込む
df = pd.read_csv('world_happiness_report.csv')

# データの最初の5行を表示
df.head()
```

```python
# データの形状を確認（行数、列数）
print(f"データの形状: {df.shape}")

# データの基本情報
df.info()
```

```python
# 基本統計量を表示
df.describe()
```

データの読み込みが終わったのでこれからデータの分析をします　\
まずはEconomy (GDP per Capita)とGenerosityのトップ１０を表示します

```python
# Economy (GDP per Capita) の高い順に10個の国を表示
print("【Economy (GDP per Capita) の高い順 TOP10】")
economy_top10 = df.nlargest(10, 'Economy (GDP per Capita)')[['Country', 'Economy (GDP per Capita)']]
print(economy_top10.to_string(index=False))

print("\n【Generosity の高い順 TOP10】")
# Generosity の高い順に10個の国を表示
generosity_top10 = df.nlargest(10, 'Generosity')[['Country', 'Generosity']]
print(generosity_top10.to_string(index=False))
```

年代を絞らないとほとんど同じ国が出てしまいました。　\
次は２０１５年の絞ってみてみます。

```python
# 2015年のデータのみをフィルタリング
df_2015 = df[df['year'] == 2015]

print("=" * 60)
print("【2015年データのみ】")
print("=" * 60)

print("\n【Economy (GDP per Capita) の高い順 TOP10】")
economy_top10_2015 = df_2015.nlargest(10, 'Economy (GDP per Capita)')[['Country', 'Economy (GDP per Capita)', 'year']]
print(economy_top10_2015.to_string(index=False))

print("\n【Generosity の高い順 TOP10】")
generosity_top10_2015 = df_2015.nlargest(10, 'Generosity')[['Country', 'Generosity', 'year']]
print(generosity_top10_2015.to_string(index=False))
```

今度はうまくできました。\
しかし、両方に名前が挙がる国がないんです\
つまり、GDPと寄付に相関関係はなかったんです\
では幸福度はどうでしょう


```python
# 2015年のHappiness Scoreが高い順に10個の国を表示
print("=" * 60)
print("【2015年 Happiness Score TOP10】")
print("=" * 60)
happiness_top10_2015 = df_2015.nlargest(10, 'Happiness Score')[['Country', 'Happiness Score', 'year']]
print(happiness_top10_2015.to_string(index=False))
```

これだけではわからないのでまとめてみました

```python
# Economy, Generosity, Happiness Scoreの3つのTOP10に共通して挙がる国を見つける
economy_countries = set(economy_top10_2015['Country'])
generosity_countries = set(generosity_top10_2015['Country'])
happiness_countries = set(happiness_top10_2015['Country'])

# Economy と Generosity の両方に挙がる国
common_economy_generosity = economy_countries & generosity_countries

# Economy と Happiness の両方に挙がる国
common_economy_happiness = economy_countries & happiness_countries

# Generosity と Happiness の両方に挙がる国
common_generosity_happiness = generosity_countries & happiness_countries

# 3つすべてに挙がる国
common_all_three = economy_countries & generosity_countries & happiness_countries

print("=" * 60)
print("【2015年 Economy と Generosity の両方に挙がる国】")
print("=" * 60)
if common_economy_generosity:
    print(f"{', '.join(sorted(common_economy_generosity))}")
else:
    print("共通の国はありません")

print("\n" + "=" * 60)
print("【2015年 Economy と Happiness Score の両方に挙がる国】")
print("=" * 60)
if common_economy_happiness:
    print(f"{', '.join(sorted(common_economy_happiness))}")
else:
    print("共通の国はありません")

print("\n" + "=" * 60)
print("【2015年 Generosity と Happiness Score の両方に挙がる国】")
print("=" * 60)
if common_generosity_happiness:
    print(f"{', '.join(sorted(common_generosity_happiness))}")
else:
    print("共通の国はありません")

print("\n" + "=" * 60)
print("【2015年 Economy, Generosity, Happiness Score の3つすべてに挙がる国】")
print("=" * 60)
if common_all_three:
    print(f"{', '.join(sorted(common_all_three))}")
else:
    print("共通の国はありません")
```

**結果** Economy と Generosity のにはあまり関係がなく、Economy と Happiness Score、Generosity と Happiness Score　にはすこし関係があるという結果になりました。
